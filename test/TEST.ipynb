named entity recognitions:

```python
def show_ents(doc):
  if doc.ents:
    for ent in doc.ents:
      print(ent.text + " - " + ent.label_ + " - " + str(spacy.explain(ent.label_)))
    else:
      print("no entity found")
```

```python
doc = nlp(u"may I go to washington, DC next May to see the Washington Monument?")
```

```python
show_ents(doc)
```

```python
 import numpy as np
 import pandas as pd
```

train a model that recognize text messages spam or ham:

```python
df = pd.read_csv("NLP/TextFiles/smsspamcollection.tsv", sep="\t")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
len(df)
```

```python
df["label"].unique()
```

```python
df["label"].value_counts()
```

text message lengths:

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.xscale('log')
bins = 1.15**(np.arange(0,50))
plt.hist(df[df['label']=='ham']['length'],bins=bins,alpha=0.8)
plt.hist(df[df['label']=='spam']['length'],bins=bins,alpha=0.8)
plt.legend(('ham','spam'))
plt.show()
```

punctiation lengths:

```python
plt.xscale('log')
bins = 1.5**(np.arange(0,15))
plt.hist(df[df['label']=='ham']['punct'],bins=bins,alpha=0.8)
plt.hist(df[df['label']=='spam']['punct'],bins=bins,alpha=0.8)
plt.legend(('ham','spam'))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# X feature data
X = df[['length','punct']]

# y is our label
y = df["label"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression(solver="lbfgs")
```

```python
lr_model.fit(X_train, y_train)
```

```python
from sklearn import metrics

# Create a prediction set:
predictions = lr_model.predict(X_test)

# Print a confusion matrix
print(metrics.confusion_matrix(y_test,predictions))
```

confusion matrix:

```python
df = pd.DataFrame(metrics.confusion_matrix(y_test,predictions), index=['ham','spam'], columns=['ham','spam'])
df
```

classification report:

```python
print(metrics.classification_report(y_test,predictions))
```

accuracy:

```python
print(metrics.accuracy_score(y_test,predictions))
```

unsuccessful detecting spams, try another model:



```python
from sklearn.naive_bayes import MultinomialNB

nb_model = MultinomialNB()

nb_model.fit(X_train, y_train)
```

```python
predictions = nb_model.predict(X_test)
print(metrics.confusion_matrix(y_test,predictions))
```

```python
print(metrics.classification_report(y_test,predictions))
```

unsuccessful detecting spams, try another model:

```python
from sklearn.svm import SVC # import the model
svc_model = SVC(gamma='auto') # create the model
svc_model.fit(X_train,y_train) # fit the model
```

```python
predictions = svc_model.predict(X_test) #predict
print(metrics.confusion_matrix(y_test,predictions)) # evaluate predections

#following predictions looks solid:
```

```python
print(metrics.classification_report(y_test,predictions))
```

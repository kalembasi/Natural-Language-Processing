```python
import spacy
import nltk
```

import english library:

```python
nlp = spacy.load("en_core_web_sm")
```

create nlp object:

```python
doc = nlp(u"Tesla is looking at buying U.S. startup for $6 million")
```

tokenization:

```python
for token in doc:
  print(token.text)
```

print part of speech (pos)

```python
for token in doc:
  print(token.text, token.pos, token.pos_)
```

```python
doc2 = nlp(u"Tesla isn't looking any startups anymore.")
```

```python
for token in doc2:
  print(token.text, token.pos, token.pos_, token.dep_)
```

may indexing doc object:

```python
doc2[0]
```

```python
doc3 = nlp(u"Apple to build a Honk Kong factory for $6 million.")
```

```python
for token in doc3:
  print(token.text, end=" | ")
```

```python
for entity in doc3.ents:
  print(entity)
  print(entity.label_)
  print("\n")
```

```python
from spacy import displacy
```

```python
displacy.render(doc4, style="ent", jupyter=True)
```

stemming:

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
words = ["run", "runner", "ran", "runs", "easily", "fairly"]
```

```python
for word in words:
  print(word + "------->" + p_stemmer.stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
s_stemmer = SnowballStemmer(language="english")
```

```python
words = ["run", "runner", "ran", "runs", "easily", "fairly"]
```

```python
for word in words:
  print(word + "------->" + s_stemmer.stem(word))
```

lemmatization:

```python
doc5 = nlp("I am runner running in a race because I love to run since I ran today")
```

```python
for token in doc5:
  print(token.text, "\t", token.pos_, "\t", token.lemma, "\t", token.lemma_)
```
